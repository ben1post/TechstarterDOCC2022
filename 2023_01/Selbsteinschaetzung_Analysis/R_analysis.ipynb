# Clean Up Data

```r
clean_colnames <- data %>%
    colnames() %>%
    sub("\\.\\."," - ",.) %>%
    gsub("\\."," ",.)


clean_colnames
```

# pivot data for analysis

```r
#colnames(data)
```

```r
data_pivot <- data %>% 
    select(-Zeitstempel, -Möchtest.Du.etwas.ergänzen.) %>% 
    gather(Variable, Antwort, -Dein.Name)
```

```r
data_pivot$Variable <- sub("\\.\\."," - ",data_pivot$Variable)
data_pivot$Variable <- gsub("\\."," ",data_pivot$Variable)
```

```r
data_pivot_groups <- data_pivot %>% separate(Variable, c("Gruppe", "Thema"), sep="-")
```

# Now let's get actual values to work with:
- convert responses to categorical variables or numeric

```r
levels(as.factor(data_pivot_groups$Antwort))
```

```r
unique(data_pivot_groups$Gruppe)
```

```r
data_num <- data_pivot_groups %>% 
    mutate(txt_Antwort = Antwort) %>%
    # check if checked all options:
    mutate(across('Antwort', str_replace, 'kann ich nicht, Kann ich im Ansatz, Kann ich sicher', 'NA')) %>%
    # check if checked 2:
    mutate(across('Antwort', str_replace, 'kann ich nicht, Kann ich im Ansatz', '1')) %>%
    mutate(across('Antwort', str_replace, 'Kann ich im Ansatz, Kann ich sicher', '3')) %>%
    mutate(across('Antwort', str_replace, 'kann ich nicht, Kann ich sicher', 'NA')) %>%
    # check single choices:
    mutate(across('Antwort', str_replace, 'kann ich nicht', '0')) %>%
    mutate(across('Antwort', str_replace, 'Kann ich im Ansatz', '2')) %>%
    mutate(across('Antwort', str_replace, 'Kann ich sicher', '4')) %>%
    mutate(num_Antwort = as.numeric(Antwort), 
           Gruppe = factor(Gruppe, levels=rev(unique(Gruppe))), 
           Thema = factor(Thema, levels=unique(Thema)),
          Name = factor(Dein.Name, levels=rev(unique(Dein.Name)))) %>%
    select(-Antwort, -Dein.Name)

data_num_norev <- data_pivot_groups %>% 
    mutate(txt_Antwort = Antwort) %>%
    # check if checked all options:
    mutate(across('Antwort', str_replace, 'kann ich nicht, Kann ich im Ansatz, Kann ich sicher', 'NA')) %>%
    # check if checked 2:
    mutate(across('Antwort', str_replace, 'kann ich nicht, Kann ich im Ansatz', '1')) %>%
    mutate(across('Antwort', str_replace, 'Kann ich im Ansatz, Kann ich sicher', '3')) %>%
    mutate(across('Antwort', str_replace, 'kann ich nicht, Kann ich sicher', 'NA')) %>%
    # check single choices:
    mutate(across('Antwort', str_replace, 'kann ich nicht', '0')) %>%
    mutate(across('Antwort', str_replace, 'Kann ich im Ansatz', '2')) %>%
    mutate(across('Antwort', str_replace, 'Kann ich sicher', '4')) %>%
    mutate(num_Antwort = as.numeric(Antwort), 
           Gruppe = factor(Gruppe, levels=unique(Gruppe)), 
           Thema = factor(Thema, levels=unique(Thema)),
          Name = factor(Dein.Name, levels=unique(Dein.Name))) %>%
    select(-Antwort, -Dein.Name)
```

## First let us look at the outliers and raw data

```r
levels(data_num$Gruppe)
```

```r
# first lets try and plot the raw data
data_num %>%
    select(Gruppe, Thema, num_Antwort, Name) %>%
    mutate(answer = as.factor(num_Antwort)) %>% 
    ggplot(aes(x = Gruppe)) +
    geom_bar(aes(fill = answer), position = "fill") -> p1
p1 + coord_flip() + theme_cowplot() + ggtitle("Antworten pro Themengruppe")
ggsave("01_ApT1.pdf")
```

```r
# first lets try and plot the raw data
data_num %>%
    select(Gruppe, Thema, num_Antwort, Name) %>%
    mutate(answer = as.factor(num_Antwort)) %>% 
    ggplot(aes(x = Name)) +
    geom_bar(aes(fill = answer), position = "fill") -> p1
p1 + coord_flip() + theme_cowplot() + ggtitle("Antworten pro Person")
ggsave("02_ApP1.pdf")
```

```r
data_num_norev %>%
    select(Gruppe, Thema, num_Antwort, Name) %>%
    #mutate(answer = as.factor(num_Antwort)) %>% 
group_by(Name) %>%
    summarise(Name = Name, Mean = mean(num_Antwort, na.rm=TRUE), SD = sd(num_Antwort, na.rm=TRUE)) %>%

    ggplot(aes(x = Name, y = Mean)) +
    geom_point(color="red", size=5) +
    geom_errorbar(aes(ymin=pmax(Mean-SD,0), ymax=Mean+SD), width=.1,
                 position=position_dodge(0.05)) -> p1
    #geom_bar(aes(fill = answer), position = "fill") -> p1
p1  + theme_cowplot() + ggtitle("Mittelwert Antworten pro Person") #+ ylim(0,4)
ggsave("03_meanApP1.pdf")
```

```r
data_num %>%
    select(Gruppe, Thema, num_Antwort, Name) %>%
    #mutate(answer = as.factor(num_Antwort)) %>% 
group_by(Gruppe) %>%
    summarise(Gruppe = Gruppe, Mean = mean(num_Antwort, na.rm=TRUE), SD = sd(num_Antwort, na.rm=TRUE)) %>%

    ggplot(aes(x = Gruppe, y = Mean)) +
    geom_point(color="red", size=5) +
    geom_errorbar(aes(ymin=pmax(Mean-SD,0), ymax=Mean+SD), width=.1,
                 position=position_dodge(0.05)) -> p1
    #geom_bar(aes(fill = answer), position = "fill") -> p1
p1 + coord_flip() + theme_cowplot() + ggtitle("Mittelwert Antworten pro Thema") #+ ylim(0,4)
ggsave("04_meanApT1.pdf")
```

```r
# Let's try generating a word cloud
library(wordcloud)
```

```r
meanthema <- data_num %>%
    select(Gruppe, Thema, num_Antwort, Name) %>%
    #mutate(answer = as.factor(num_Antwort)) %>% 
    group_by(Thema) %>%
    summarise(Mean = mean(num_Antwort, na.rm=TRUE))
head(meanthema)
```

```r
pdf("WC_03_positiv_Thema.pdf")
options(repr.plot.width=10, repr.plot.height=10)

set.seed(1234)
wordcloud(words = meanthema$Thema, freq = 10**meanthema$Mean, min.freq = 0.5,           
          max.words=200, random.order=FALSE, rot.per=0.35,            
          colors=brewer.pal(8, "YlGn"), scale=c(3, .2))
dev.off()
```
